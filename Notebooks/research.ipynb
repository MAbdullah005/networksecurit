```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
from NetworkSecurity.constant.traning_pipeline import FILE_NAME
```

```python
df=pd.read_csv(f'../network_data/{FILE_NAME}')
df.sample(5)
```

```python
df.shape
```

```python

# Basic info
print(df.shape)              # Rows, columns
print(df.info())             # Dtype, nulls
print(df.describe())         # Summary stats
print(df['Result'].value_counts())  # Class distribution
```

```python
print(df.isnull().sum())      # Missing values
print("if missing value has place holder")
print((df == '?').sum())      # If some fields use '?' as placeholder
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
df
```

```python

sns.countplot(df['Result'])
plt.title("Phishing vs Legitimate Count")
plt.show()
```

```python
corr = df.corr()
plt.figure(figsize=(16, 12))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
important_features = ['URL_Length', 'SSLfinal_State', 'web_traffic', 'Page_Rank']

for col in important_features:
    sns.histplot(df[col], kde=False, bins=10)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in ['URL_Length', 'web_traffic', 'Page_Rank']:
    sns.boxplot(x='Result', y=col, data=df)
    plt.title(f"{col} vs Phishing")
    plt.show()
```

```python
from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.01)
selector.fit(df.drop(columns=["Result"]))

low_variance_features = df.drop(columns=["Result"]).columns[~selector.get_support()]
print("Low variance features:", low_variance_features)
```

```python
import seaborn as sns
top_corr = df.corr()['Result'].abs().sort_values(ascending=False)[1:6].index.tolist()
sns.pairplot(df[top_corr + ['Result']], hue='Result')
```

```python
df['Result'].value_counts(normalize=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop("Result", axis=1)
y = df["Result"]

model = RandomForestClassifier()
model.fit(X, y)

importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values().plot(kind='barh', figsize=(10, 8), title="Feature Importances")
```
